```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", category=sp.sparse.SparseEfficiencyWarning)
np.set_printoptions(linewidth=140)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 5))
X = np.linspace(-1, 1, 100)

axs[0, 0].plot(X, X**2+0)
axs[0, 0].plot(X, X**2-0.1)
axs[0, 1].plot(X, X**2+0.1)
axs[1, 0].plot(X, X**2+0.2)
axs[1, 1].plot(X, X**2+0.3)

for row in axs:
    for ax in row:
        ax.set_xlim(-1, 1)
        ax.set_ylim(-.1, 1)
        ax.grid()
```

```python
ax = plt.figure().add_subplot(projection='3d')
X = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(X, X)
Z = X**2/4+Y**2/7
ax.plot_surface(X, Y, Z, rstride=5, cstride=5, cmap='viridis')
ax.set(xlim=(-1, 1), ylim=(-1, 1), zlim=(-0.1, 0.5),
    xlabel='X', ylabel='Y', zlabel='Z')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
X = np.arange(100)
ax.loglog(X, X**2, label='X**2')
ax.loglog(X, X**3, label='X**3')
ax.loglog(X, X**4, label='X**4')
ax.legend()
```
